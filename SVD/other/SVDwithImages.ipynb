```python

import matplotlib.pyplot as plt
from matplotlib.image import imread
import numpy as np

## A = imread(  '/Users/user/Desktop/dog.jpg'  )
A = imread(  '/Users/user/Desktop/jupiter.jpg'  )
```

```python

A.shape
```

```python

## mean the 3 RGBs to make grey scale

X = np.mean(A, -1)

print( X.shape )
```

```python

img = plt.imshow( X  )

img
```

```python

U, S, VT = np.linalg.svd( X, full_matrices=False)

S = np.diag( S )
```


## Approximate matrix (image) with truncated SVD for various ranks


```python

for r in (5, 20, 100):
    
    Xapprox = U[:, :r] @ S[0:r, :r] @ VT[:r, :]
    img     = plt.imshow(  Xapprox  )
    plt.show()
```

```python

plt.semilogy(  np.diag(S)  )
plt.plot(      np.cumsum(np.diag(S))  / np.sum(np.diag(S))    )
```
