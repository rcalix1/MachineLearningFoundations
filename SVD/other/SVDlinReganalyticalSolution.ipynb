```python

import numpy as np
import matplotlib.pyplot as plt
```

```python

w = 3    ## slope

x = np.arange(-2, 2, 0.25)

print( x )
print( x.shape )
```

```python

x = x.reshape( -1, 1)

print( x )
print( x.shape )
```

```python

y = w*x + np.random.randn(*x.shape)  ## add noise

print( y )
print( y.shape )
```

```python

plt.plot(x, w*x, color='k', linewidth=2, label='True Line')
```

```python

plt.plot(x, y, 'x', color='k', markersize=10, label='Noisy data')  ## noisy
```

```python

## Compute Least squares approximation with SVD

U, S, VT = np.linalg.svd(x, full_matrices=False)
```

```python

wtilde = VT.T @ np.linalg.inv( np.diag(S) ) @ U.T @ y ## Least squares fit
```

```python

wtilde
```

```python

plt.plot(x, wtilde*x, '--', 'x', color='b', linewidth=4, label='Regression  Line')  
```

```python

plt.plot(x, w*x, color='k', linewidth=2, label='True Line')
plt.plot(x, y, 'x', color='k', markersize=10, label='Noisy data')  ## noisy
plt.plot(x, wtilde*x, '--', 'x', color='b', linewidth=4, label='Regression  Line')
```
