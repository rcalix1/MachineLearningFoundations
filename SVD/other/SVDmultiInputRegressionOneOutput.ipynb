
## SVD to solve for w of multi-input and 1 output linear regression 


```python

X = np.loadtxt('/Users/user/Desktop/hald_ingredients.csv', delimiter=',')

y = np.loadtxt('/Users/user/Desktop/hald_heat.csv', delimiter=',')
```


## Solve y = Xw using SVD


```python

U, S, VT = np.linalg.svd( X, full_matrices=0)
```

```python

U
```

```python

S
```

```python

VT
```

```python

w = VT.T @ np.linalg.inv(  np.diag(S)  )  @  U.T @  y

## or

## w = np.linalg.pinv(X)  @  y     ## alternative
```

```python

plt.plot(y, color='k', linewidth=2, label='Heat Data')
plt.plot(X@w, '-o' , color='r',  label='Regression')
```
